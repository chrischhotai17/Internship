## Question 2

```python
page = requests.get("https://www.imdb.com/chart/top/")

soup = BeautifulSoup(page.content)
soup
```

```python
table = soup.find("table", {"class": "chart"})

rows = table.tbody.find_all("tr")
rows = rows[:100]

titles_list = []
year_list = []
rating_list = []


for row in rows:
#     print(row)
    titleColumn = row.find("td", {"class": "titleColumn"})
    title = titleColumn.a.text
    year = titleColumn.span.text[1:-1]
    rating = row.find("td", class_="imdbRating").strong.text
    
    titles_list.append(title)
    year_list.append(year)
    rating_list.append(rating)
    print(year, ":", title, ":", rating)
```

```python
import pandas as pd
df=pd.DataFrame({"titles":titles_list,"year":year_list,"rating":rating_list})
df
```

## i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
# Table for top 10 teams by rankings in ODI matches
url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
page = requests.get(url)
soup = BeautifulSoup(page.content)
table = soup.table
table.tbody
```

```python
countries = []
matches = []
points = []
ratings = []

# For first row (#1 team), the class names are different so have to be handled separately
countries.append(table.find(class_="rankings-block__banner--team-name").find('span', class_='u-hide-phablet').text.strip())
matches.append(table.find(class_="rankings-block__banner--matches").text.strip())
points.append(table.find(class_="rankings-block__banner--points").text.strip())
ratings.append(table.find(class_="rankings-block__banner--rating").text.strip())

countries, matches, points, ratings
```

```python
# For the other 9 rows:

for row in table.tbody.find_all('tr', class_="table-body")[:9]:
    country = row.find(class_="table-body__cell rankings-table__team").find('span', class_='u-hide-phablet').text.strip()
    match_col, points_col = row.find_all(class_="table-body__cell u-center-text")
    match = match_col.text.strip()
    point = points_col.text.strip()
    rating = table.find(class_="table-body__cell u-text-right rating").text.strip()
    
    countries.append(country)
    matches.append(match)
    points.append(point)
    ratings.append(rating)
    
countries, matches, points, ratings
```

```python
df = pd.DataFrame({"Team Name":countries, "Matches":matches, "Points":points, "Ratings":ratings})

df
```

## Question 8

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
# page content
soup=BeautifulSoup(page.content)
```

```python
name=[]
for i in soup.find_all("div", class_="restnt-info cursor"):
    name.append(i.text)
# for location
location=[]
for i in soup.find_all("div", class_="restnt-loc ellipsis"):
    location.append(i.text)
# for price and cusine
price = []
cuisine = []
for i in soup.find_all("span", class_="double-line-ellipsis"):
    price.append(i.text.split('|')[0])
    cuisine.append(i.text.split('|')[1])
    
# for rating
rating=[]
for i in soup.find_all("div", class_="restnt-rating rating-3"):
    rating.append(i.text)
for i in soup.find_all("div", class_="restnt-rating rating-4"):
    rating.append(i.text)

# for images
images = []
for i in soup.find_all("img", class_="no-img"):
    images.append(i['data-src'])
```

```python
print(len(name), len(location), len(price), len(cuisine), len(rating), len(images))
```

```python
# Make data frame
DineOut=pd.DataFrame({})
DineOut['Restaurant Name']=name
DineOut['Location']=location
DineOut['Price']=price 
DineOut['Cuisine']=cuisine  
DineOut['Rating']=rating  
DineOut['IMAGES']=images
DineOut
```

## Question 9

```python
url = 'https://en.tutiempo.net/delhi.html?data=last-24-hours'
page = requests.get(url)
page
```

```python
htmlContent= page.content
soup= BeautifulSoup(htmlContent,'html.parser')
```

```python
#creating empty list
Temperature = []
Wind_km_hr = []
Humidity= []
Pressure_hPA = []
Weather_condition = []
Hour=[]
```

```python
temp = soup.find_all ('td',class_='t Temp') #fetching temperature
wind = soup.find_all ('td',class_='wind') #fetching wind
humid = soup.find_all ('td',class_='hr') #fetching humidity
pressure = soup.find_all ('td',class_='prob') #fetching pressure

for i in range(0,24):
    Temperature.append(temp[i].get_text().replace("\n",""))
    Wind_km_hr.append(wind[i].get_text().replace("\n",""))
    Humidity.append(humid[i].get_text().replace("\n",""))
    Pressure_hPA.append(pressure[i].get_text().replace("\n",""))
    
for i in soup.find_all ('td',class_='t Temp')[0:24]: 
    if i.previous_sibling.previous_sibling is not None:
        Hour.append(i.previous_sibling.previous_sibling.text)
    else:
        Hour.append(' ')

for i in soup.find_all ('td',class_='t Temp')[0:24]: 
    if i.previous_sibling is not None:
        Weather_condition.append(i.previous_sibling.text)
    else:
        Weather_condition.append(' ')
```

```python
data = list(zip(Hour,Temperature,Wind_km_hr,Humidity,Pressure_hPA,Weather_condition))
df = pd.DataFrame(data,columns=['Hour','Temperature','Wind_km_hr','Humidity','Pressure_hPA','Weather_condition'])
df
```

## Question 10

```python
url = 'https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/'
response = requests.get(url) # in order to work with html we need to get it as string
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
data = soup.find_all('div', class_="blog--single__content column--3-4 u-spacing-third")
temp=[]
for i in data:
    for j in i.find_all('p'):
        t = j.text.replace('\n', '') # only extracting data
        temp.append(t)
name=[]
desc=[]
for i in range(1,30,3):
    name.append(temp[i])
for i in range(2,30,3):
    desc.append(temp[i])
temp = [] # scraping all image url's
for i in soup.find_all('img'):
    temp.append(i['src'])
img_url=[]
for i in range(6,26,2):
    img_url.append(temp[i])
```

```python
df = pd.DataFrame()
df['Monument'] = name
df['Description'] = desc
df['Image_URL']= img_url
df
```
